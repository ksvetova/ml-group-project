```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# PREPARING THE DATA
```

```python
npf_train = pd.read_csv("npf_train.csv")
npf_train = npf_train.drop("partlybad",axis=1)
npf_train
```

```python
# COPYING MODIFICATIONS FROM e1t1

npf_train = npf_train.set_index("date")
## Tell that "class4" is categorical variable. (R does this automatically.)
npf_train["class4"] = npf_train["class4"].astype("category")

## Here date column was converted to index and we do not need to get rid of it.
npf_train = npf_train.drop("id",axis=1)

## If you don't use dtype="object" array will cut strings...
class2 = np.array(["event"]*npf_train.shape[0],dtype="object")
class2[npf_train["class4"]=="nonevent"] = "nonevent"
npf_train["class2"] = class2
npf_train["class2"] = npf_train["class2"].astype("category")
```

```python
npf_train
```

```python
# BUILDING THE CLASSIFIERS
```

```python
# I will just train a basic model fpr each model family,
# compare the results and choose the one with the best accuracy/lowest test error 
```

```python
# creating a df to store the models and results in
models = pd.DataFrame()
models["model"] = ["LDA", "QDA", "log regr", "nB", "KNN k=10", "KNN k = 20", "KNN k = 30", "decision tree", "random forest"]
models = models.set_index("model")
models

# seperating X and y from the training data # !!!!!!! we predict class 2 for now !!!!!!!


# reformating X to numerics
for i in range(1,101): # 101 because we have 102 columns
    npf_train.iloc[:,i] = pd.to_numeric(list(npf_train.iloc[:,i]))

## REDUCING the set to get a training and a test set (360 ~ 80% of 458)
X = npf_train.iloc[:360, 1:-1]
#y = [1 if i == "event" else 0 for i in npf_train.iloc[:360,-1]] # 1: event, 0: nonevent
y = npf_train.iloc[:360, -1]

# imports for the models
import sklearn
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# LDA
model_lda = LinearDiscriminantAnalysis()
t_model_lda = model_lda.fit(X, y)

# QDA
model_qda = QuadraticDiscriminantAnalysis()
t_model_qda = model_qda.fit(X, y)

# logistic regression
model_lr = LogisticRegression()
t_model_lr = model_lr.fit(X, y)

# naive Bayes
model_nb = GaussianNB()
t_model_nb = model_nb.fit(X, y)

# KNN
model_knn10 = KNeighborsClassifier(n_neighbors = 10)
t_model_knn10 = model_knn10.fit(X, y)
model_knn20 = KNeighborsClassifier(n_neighbors = 20)
t_model_knn20 = model_knn20.fit(X, y)
model_knn30 = KNeighborsClassifier(n_neighbors = 30)
t_model_knn30 = model_knn30.fit(X, y)

# decision tree
model_dt = DecisionTreeClassifier()
t_model_dt = model_dt.fit(X,y)

# random forest
model_rf = RandomForestClassifier()
t_model_rf = model_rf.fit(X,y)

# add trained models to df
models["trained_models"] = [t_model_lda, t_model_qda, t_model_lr, t_model_nb, t_model_knn10, t_model_knn20, t_model_knn30, t_model_dt, t_model_rf]
models

# show choice of category
# show probability of belonging to event
# show test error
```

```python
# our test set:
X_test = npf_train.iloc[360:, 1:-1]
y_test = npf_train.iloc[360:, -1]
# computing the scores
scores = []
for i in models.iloc[:, 0]:
    scores.append(i.score(X_test, y_test))
    
models["test score"] = scores
models


# --> logistic regression! (when random forest is not considered)
```
